```python
import numpy as np
import matplotlib.pyplot as plt

# Define Years (e.g., 2015 to 2024)
years = np.arange(2014, 2024)  # 10 years

# Sample Growth Rates for Each Sector (Replace with real data)
primary_sector = [16.79,16.17,16.36,16.56,16.03,16.76,18.67,17.37,16.64,16]  # Growth rate between 1% to 5%
secondary_sector = [27.66,27.35,26.62,26.5,26.41,24.59,25.12,26.47,25.28,25.03]# Growth rate between 5% to 15%
tertiary_sector =   [47.82,47.78,47.75,47.67,48.43,50.85,57.93,47.85,49.58,49.58]# Growth rate between 5% to 12%

x = np.arange(len(years))  # X-axis positions for bars

# Create figure and axis
fig, ax = plt.subplots(figsize=(12, 6))

# Bar Width (Adjusted for multiple years)
bar_width = 0.3

# Plot bars for each sector
ax.bar(x - bar_width, primary_sector, bar_width, color='darkgreen', label="Agriculture Sector")
ax.bar(x, secondary_sector, bar_width, color='goldenrod', label="Industry Sector")
ax.bar(x + bar_width, tertiary_sector, bar_width, color='navy', label="Services Sector")

# Labels and title
ax.set_title("GDP Growth Rates (%) of Different Sectors", fontsize=14, fontweight='bold')
ax.set_xlabel("Years", fontsize=12)
ax.set_ylabel("Growth Rate (%)", fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(years, rotation=45)

# Grid and legend
ax.grid(axis='y', linestyle="--", alpha=0.5)
ax.legend()
# Show plot
plt.show()
```
